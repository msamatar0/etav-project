```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("planets")
x ="distance"
y = "mass"
title="Planet"
```

```python
df
```

```python
df.drop_duplicates("mass")
```

```python
col = "distance"
max = df[col].max()
min = df[col].min()
avg = df[col].mean()
print(f"Max: {max}\nMin: {min}\nAverage: {avg}")
```

```python
plt.plot(df.index, df[y])
plt.grid()
plt.xlabel(x)
plt.ylabel(y)
plt.title(title)
plt.show()
```

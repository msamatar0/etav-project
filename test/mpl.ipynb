```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(-5, 5.1, .1)
e_x = np.exp(x)
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)

df = pd.DataFrame({"e^x": e_x, "sin(x)": sin_x, "cos(x)": cos_x, "tan(x)": tan_x}, index=pd.Series(x, name="x"))
print(df)
df.plot()
```

```python
df.plot(x="e^x",y=["sin(x)", "cos(x)"])
```

```python
df.plot(x="sin(x)",y="cos(x)", kind="scatter")
```

```python
plt.plot(df.index, df["sin(x)"])
plt.grid()
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("trig - sin of x")
plt.show()
```

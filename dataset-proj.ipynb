```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = pd.read_csv("games.csv", index_col=0)
df.index.name ="id"
df.drop("Times Listed", axis=1, inplace=True)
```

```python
df
```

```python
col = "Rating"
max = df[col].max()
min = df[col].min()
avg = df[col].mean()
print(f"Max: {max}\nMin: {min}\nAverage: {avg}")
```

```python
x ="Title"
y = "Plays"
title = x + " vs " + y.rstrip('s')

top_df = df.sort_values(y, ascending=False).head(5)

plt.bar(top_df[x]., top_df[y])
#plt.xlim(1,5)
#plt.ylim(4.5,5)
plt.xticks(rotation=45)
plt.grid()
plt.xlabel(x)
plt.ylabel(y)
plt.title(title)
plt.show()
```

```python
plt.close("all")
```

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns, re
```

```python
df = pd.read_csv("games.csv", index_col=0).head(100)#.sort_values("Rating", ascending=False).head(20)
df.index.name ="id"
df.drop("Times Listed", axis=1, inplace=True)
```

```python
df
```

```python
col = "Rating"
max = df[col].max()
min = df[col].min()
avg = df[col].mean()
print(f"Max: {max}\nMin: {min}\nAverage: {avg}")
```

```python
x ="Title"
y = "Rating"
title = x + " vs " + y.rstrip('s')

plt.plot(df.index, df[y])
plt.xticks(rotation=45)
plt.grid()
plt.xlabel(x)
plt.ylabel(y)
plt.title(title)
plt.show()
```

```python
play_series = pd.Series(df["Plays"])

for s in df["Plays"].values:
    if re.match("\d\.\dK", s):
        n = float("" + s[0] + s[1] + s[2])
    elif re.match("\d{2}K", s):
        n = float("" + s[0] + s[1])
    play_series.add(n)
print(play_series)
```
